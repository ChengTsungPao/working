```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors

import torch
import torch.nn as nn
import torch.optim as optim
import torch.utils.data as Data
from torch.autograd import Variable

from sklearn.cluster import KMeans
```

```python
class RBM(nn.Module):
    def __init__(self, visible, hidden, k=5, device="cuda:0"):
        super(RBM, self).__init__()
        
        self.weight = nn.Parameter(torch.randn(hidden, visible).cuda())
        self.vis_bias = nn.Parameter(torch.zeros(visible).cuda())
        self.hid_bias = nn.Parameter(torch.zeros(hidden).cuda())
        self.k = k
        self.device = device
        
        if device != "cuda:0":
            self.weight = nn.Parameter(torch.randn(hidden, visible))
            self.vis_bias = nn.Parameter(torch.zeros(visible))
            self.hid_bias = nn.Parameter(torch.zeros(hidden))

    def sampling(self, p):
        if self.device != "cuda:0":
            return nn.functional.relu(torch.sign(p - torch.rand(p.size())))
        else:
            return nn.functional.relu(torch.sign(p - torch.rand(p.size()).cuda()))
        
    def visible_to_hidden(self, visible):
        x = torch.sigmoid(nn.functional.linear(visible, self.weight, self.hid_bias))
        sample = self.sampling(x)
        return x, sample
    
    def hidden_to_visible(self, hidden):
        x = torch.sigmoid(nn.functional.linear(hidden, self.weight.t(), self.vis_bias))
        sample = self.sampling(x)
        return x, sample
        
    def forward(self, Input):
        pre_h1, h1 = self.visible_to_hidden(Input)
        
        h_ = h1
        for _ in range(self.k):
            pre_v_,v_ = self.hidden_to_visible(h_)
            pre_h_,h_ = self.visible_to_hidden(v_)
        
        return Input, v_, h_
    
    def free_energy(self, v):
        vbias_term = v.mv(self.vis_bias)
        wx_b = nn.functional.linear(v, self.weight, self.hid_bias)
        hidden_term = wx_b.exp().add(1).log().sum(1)
        return (-hidden_term - vbias_term).mean()
    
    def loss(self, v1, v2):
        return self.free_energy(v1) - self.free_energy(v2)
```

```python
file_name = "BA_matrix_train,L=10,mu=[1,3],delta=1.npz"

npzfile = np.load('data/' + file_name)
X = npzfile['arr_0']
Y = npzfile['arr_1']
mu = npzfile['arr_2']
npzfile.close()
```

```python
sets, L, _ = X.shape

for i in range(sets):
    X[i, :, :] = X[i, :, :]-np.min(X[i, :, :])
    X[i, :, :] = X[i, :, :]/np.max(X[i, :, :])

X = np.reshape(X, [sets, L**2])
np.random.shuffle(X)
Y = np.reshape(Y, [sets, 1])

validation_split = 0.3

train_data = Data.TensorDataset(torch.from_numpy(X[:int(sets*(1-validation_split)), :]), 
                                torch.from_numpy(Y[:int(sets*(1-validation_split)), :])
                               )
validation_data = Data.TensorDataset(torch.from_numpy(X[int(sets*(1-validation_split)):, :]), 
                                     torch.from_numpy(Y[int(sets*(1-validation_split)):, :])
                                    )
```

```python
rbm = RBM(L**2, 200)
device = torch.device("cuda:0")
rbm = rbm.to(device)
```

```python
loss_ = []
vloss_ = []

Epochs = 500
Batch_Size = 64
Rate = 1.
gamma = 0.1

Optimizer = optim.SGD(rbm.parameters(), Rate)
Loss = rbm.loss

print("train on " + str(int(sets*(1-validation_split))) + \
      " samples, validate on " + str(int(sets*validation_split)) + " samples")
for epoch in range(Epochs):
    train_loader = Data.DataLoader(dataset=train_data, batch_size=Batch_Size, shuffle=True)
    validation_loader = Data.DataLoader(dataset=validation_data, batch_size=Batch_Size, shuffle=True)
    for _, (data, target) in enumerate(train_loader):
        data = data.view(-1, L**2).cuda().float()
        x = Variable(data)
        x = x.bernoulli()
        x = x.to(device)
        
        v1, v2, _ = rbm(x)
        loss = Loss(v1, v2)
        Optimizer.zero_grad()
        loss.backward()
        Optimizer.step()
        
    for _, (data, target) in enumerate(validation_loader):
        data = data.view(-1, L**2).cuda().float()
        x = Variable(data)
        x = x.bernoulli()
        x = x.to(device)

        v1, v2, _ = rbm(x)
        validation_loss = Loss(v1, v2)
    
    vloss_.append(loss.data.cpu().numpy())
    loss_.append(loss.data.cpu().numpy())
    if epoch % 100 == 0:
        Rate *= gamma
        Optimizer = optim.SGD(rbm.parameters(), Rate)
    print("epoch " + str(epoch+1) + "/" + str(Epochs) +
          "\nloss " + str(loss.data.cpu().numpy()) + ", validation loss " + str(validation_loss.data.cpu().numpy()))

    torch.cuda.empty_cache()
```

```python
plt.figure()
plt.plot([i+1 for i in range(Epochs)], loss_, label="loss")
plt.plot([i+1 for i in range(Epochs)], vloss_, label="validation loss")
plt.legend()
plt.show()
```

```python
model_name = "BA_train_model,L=10,mu=[0,10],delta=1.pth"
torch.save(rbm.state_dict(),"model/" + model_name)
```
