```python
import io 
import urllib
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_image_from_url(image_url, size = (200, 200)):
    file_descriptor = urllib.request.urlopen ( image_url )
    image_file = io.BytesIO( file_descriptor.read ())
    image = Image.open( image_file )
    #img_shape = (image.size[1], image.size[0])
    img_color = image.resize (size, 1)
    img_grey = img_color.convert('L')
    img = np.array(img_grey, dtype = np.float )
    return img
```

```python
plt.style.use('classic')
img = get_image_from_url( image_url='https://testimages.juliaimages.org/thumbnails/peppers_gray.png', size = (256, 300))

plt.figure(figsize=(9, 6))
plt.imshow(img, cmap='gray')
plt.show()
```

```python
U, D, V = np.linalg.svd(img)
print('image shape: %s U shape: %s  D shape: %s  V shape: %s'%(img.shape, U.shape , D.shape, V.shape))
```

```python
for i in [5, 10, 15, 20, 30, 50]:
    reconstimg = np.matrix(U[:, :i]) * np.diag(D[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```
